```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import os
import csv
```

```python
basePath = "CSV_DataFiles/"
resultPath = "5%-Standard_Scaled_CSV_Files/"
filePaths = os.listdir(basePath)
```

```python
mean_all_files = []
count = 0
for curFilePath in filePaths:
    curFileData = []
    with open(basePath + curFilePath, "r") as file:
        reader = csv.reader(file)
        for row in reader:
            curFileData.append(row)
    fivePer = int(0.05 * len(curFileData))
    dataToFit = curFileData[:fivePer]
    dataLeft = curFileData[fivePer:]
    scaler = StandardScaler()
    scaler.fit(dataToFit)
    mean_all_files.append(scaler.mean_)
    transformedData1 = scaler.transform(dataToFit)
    transformedData2 = scaler.transform(dataLeft)
    with open(resultPath + curFilePath, "w", newline='') as f:
        writer = csv.writer(f)
        for row1 in transformedData1:
            writer.writerow(row1)
        for row2 in transformedData2:
            writer.writerow(row2)
    print("File : ", count, " -> Done")
    count += 1
```

```python
with open("5%_StdMean_All.csv", "w", newline='') as f:
    writer = csv.writer(f)
    for row in mean_all_files:
        writer.writerow(row)
```
