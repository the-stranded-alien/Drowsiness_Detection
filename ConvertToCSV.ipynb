```python
import numpy as np
import pandas as pd
import os
```

```python
folders = os.listdir("DataFiles/")
```

```python
folders = folders[1:-1]
print(folders)
```

```python
dataFilePaths = []

for folds in folders:
    path = "DataFiles/" + folds
    Files = os.listdir(path)
    for file in Files:
        dataFilePaths.append(path + "/" + file)
```

```python
data = []
counter = 0
for filePath in dataFilePaths:
    lineCount = 0
    with open(filePath) as file:
        currentFileData = file.readlines()
        fileData = []
        for lines in currentFileData:
            fileData.append(lines)
            lineCount +=1
    data.append(fileData)
    counter += 1
    print("File ", counter, " Done. Lines : ", lineCount)
```

```python
import csv
data1 = data[:90]
data2 = data[90:]
print(len(data1), len(data2))
fileNum = 0
```

```python
for currentFile in data1:
    file = []
    for eachLine in currentFile:
        eachLine = eachLine.split(",")
        eachLine = eachLine[:-1]
        curLine = []
        for eachPoint in eachLine:
            curLine.append(int(eachPoint))
        file.append(curLine)
    with open("CSV_DataFiles/" + str(int(fileNum/3) + 1) + "-" + str(int(fileNum%3)*5) + ".csv", "w", newline='') as f:
        writer = csv.writer(f)
        for row in file:
            writer.writerow(row)
    print("File ", fileNum, " -> Done.")
    fileNum += 1
```
