```python
import os
import csv
import numpy as np
import pandas as pd
```

```python
basePath = "CSV_DataFiles/"
filePaths = os.listdir(basePath)
f1 = filePaths[:90]
f2 = filePaths[90:]
```

```python
count = 0
for curFile in f1:
    newFile = []
    with open(basePath + curFile, "r") as file:
        reader = csv.reader(file)
        for row in reader:
            newLine = []
            x_row = []
            y_row = []
            for i in range(len(row)):
                if(i % 2 == 0):
                    x_row.append(row[i])
                else:
                    y_row.append(row[i])
            xArray = np.array(x_row)
            yArray = np.array(y_row)
            xMin = min(xArray)
            xMax = max(xArray)
            yMin = min(yArray)
            yMax = max(yArray)
            newLine.append(xMin)
            newLine.append(xMax)
            newLine.append(yMin)
            newLine.append(yMax)
            newLine.append(x_row)
            newLine.append(y_row)
            newFile.append(newLine)
    with open("MinMaxCSV_DataFiles/" + str(curFile), "w", newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["xMin","xMax","yMin","yMax","68_X_Points","68_Y_Points"])
        for row in newFile:
            writer.writerow(row)
    count += 1
    print("File : ", count, " -> Done")
```
