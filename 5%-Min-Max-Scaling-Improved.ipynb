```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import os
import csv
import pandas as pd
```

```python
basePath = "MinMaxCSV_DataFiles/"
resultPath = "5%-MinMax_Scaled_CSV_Files_Imporved/"

filePaths = os.listdir(basePath)
```

```python
for file in filePaths:
    
    # Creating dataframes for each file and separating them into two.
    dataFrame = pd.read_csv(basePath + file)
    fivePer = int(0.05 * len(dataFrame))
    df_fivePer = pd.DataFrame(dataFrame[:fivePer])
    df_left = pd.DataFrame(dataFrame[fivePer:])
    df_left.reset_index(inplace=True, drop=True)
    
    # Finding min & max, X and Y values for first 5%. 
    X_MIN_5 = min(df_fivePer["xMin"])
    X_MAX_5 = max(df_fivePer["xMax"])
    Y_MIN_5 = min(df_fivePer["yMin"])
    Y_MAX_5 = max(df_fivePer["yMax"])
   
    all_Scaled_X_values = []
    all_Scaled_Y_values = []

    # Scaling X values of df_fivePer
    for st1 in df_fivePer["68_X_Points"]: 
        xVals = []
        xScaled = []
        st1 = st1[1:-1].split(",")
        for num in st1:
            N = num[1:-1]
            if N[0] == "'":
                xVals.append(int(N[1:]))
            else:
                xVals.append(int(N))
        
        for vals in xVals:
            scaled = ((vals - X_MIN_5)/(X_MAX_5 - X_MIN_5))
            xScaled.append(scaled)
        all_Scaled_X_values.append(xScaled)
        
    # Scaling Y values of df_fivePer
    for st2 in df_fivePer["68_Y_Points"]: 
        yVals = []
        yScaled = []
        st2 = st2[1:-1].split(",")
        for num in st2:
            N = num[1:-1]
            if N[0] == "'":
                yVals.append(int(N[1:]))
            else:
                yVals.append(int(N))
        
        for vals in yVals:
            scaled = ((vals - Y_MIN_5)/(Y_MAX_5 - Y_MIN_5))
            yScaled.append(scaled)
        all_Scaled_Y_values.append(yScaled)
    
    # Scaling X values of df_left
    for st3 in df_left["68_X_Points"]: 
        xVals = []
        xScaled = []
        st3 = st3[1:-1].split(",")
        for num in st3:
            N = num[1:-1]
            if N[0] == "'":
                xVals.append(int(N[1:]))
            else:
                xVals.append(int(N))
        
        for vals in xVals:
            scaled = ((vals - X_MIN_5)/(X_MAX_5 - X_MIN_5))
            xScaled.append(scaled)
        all_Scaled_X_values.append(xScaled)
        
    # Scaling Y values of df_left
    for st4 in df_left["68_Y_Points"]: 
        yVals = []
        yScaled = []
        st4 = st4[1:-1].split(",")
        for num in st4:
            N = num[1:-1]
            if N[0] == "'":
                yVals.append(int(N[1:]))
            else:
                yVals.append(int(N))
        
        for vals in yVals:
            scaled = ((vals - Y_MIN_5)/(Y_MAX_5 - Y_MIN_5))
            yScaled.append(scaled)
        all_Scaled_Y_values.append(yScaled)
    
    # Printing Scaled Values for Current File
    with open(resultPath+file,"w",newline='') as nf:
        writer = csv.writer(nf)
        writer.writerow(["X_Scaled_Values","Y_Scaled_Values"])
        for i in range(len(all_Scaled_X_values)):
            l1 = all_Scaled_X_values[i]
            l2 = all_Scaled_Y_values[i]
            nl = [l1,l2]
            writer.writerow(nl)
            
    # Loop Summary 
    print(file, " : Done.")
```
